```python
TARGET_DTE = 60
EXIT_DTE = 5
HEDGE_THRESHOLD = 5
CAPITAL = 20000
MAX_IV_THRESHOLD = 25.0
#PARAMS
```

```python
#Ensure sort
spy_df.index = pd.to_datetime(spy_df.index)
unique_dates = sorted(spy_df.index.unique())
```

```python
# func to get close straddle
def get_atm_straddle(daily_df, target_dte):
    """
    Finds the Call and Put option closest to At-The-Money (ATM) 
    with DTE closest to the target.
    """
    daily_df = daily_df.copy().reset_index(drop=True)
    
    # get within a week
    daily_df['dte'] = daily_df['dte'].astype(float)
    valid_expires = daily_df[(daily_df['dte'] >= target_dte - 7) & 
                             (daily_df['dte'] <= target_dte + 7)]
    
    if valid_expires.empty:
        return None, None
    
    # close to target
    best_dte_idx = valid_expires['dte'].sub(target_dte).abs().idxmin()
    target_expire_date = valid_expires.loc[best_dte_idx, 'expire_date']
    
    # 3. option chain for exp
    chain = daily_df[daily_df['expire_date'] == target_expire_date].copy()
    
    
    chain = chain.reset_index(drop=True)
    
    current_price = float(chain['underlying_last'].iloc[0])
    
    # Calculate distance
    chain['dist'] = chain['strike'].astype(float).sub(current_price).abs()
    
    # Find the row with min distance
    best_strike_idx = chain['dist'].idxmin()
    atm_row = chain.loc[best_strike_idx]
    
    return atm_row, current_price
```

```python
cash = CAPITAL
stock_inventory = 0
portfolio_value_history = []
stock_pnl_history = []
option_pnl_history = []

current_option = None
entry_date = None

print(f"Starting Backtest on {len(unique_dates)} trading days...")

for date in unique_dates:
    daily_data = spy_df.loc[date]
    
    if isinstance(daily_data, pd.Series):
        daily_data = daily_data.to_frame().T
        
    current_price = float(daily_data['underlying_last'].iloc[0])

    opt_row = None 

    # Enter a position if we haven't
    if current_option is None:
        atm_row, u_price = get_atm_straddle(daily_data, TARGET_DTE)
        
        if atm_row is not None:

            # calculate iv
            current_iv = (float(atm_row['c_iv']) + float(atm_row['p_iv'])) / 2
            
            # skip if too expensive
            if current_iv > MAX_IV_THRESHOLD:
                print(f"[{date.date()}] SKIPPED: IV is {current_iv:.2f} (Too Expensive!)")
                continue 
            

            # BUY STRADDLE (1 Call + 1 Put)
            cost = (float(atm_row['c_ask']) + float(atm_row['p_ask'])) * 100
            cash -= cost

            
            current_option = {
                'strike': atm_row['strike'],
                'expire_date': atm_row['expire_date'],
                'quantity': 1 
            }
            
            opt_row = atm_row 

            # hedge
            net_delta = (float(atm_row['c_delta']) + float(atm_row['p_delta'])) * 100
            shares_to_buy = -1 * (net_delta + stock_inventory)
            
            cash -= (shares_to_buy * current_price)
            stock_inventory += shares_to_buy
            
            print(f"[{date.date()}] ENTER: Strike {atm_row['strike']} | Hedge: {shares_to_buy:.1f} shares")

    # Manage curr position
    else:
        # refind our options
        options_today = daily_data[
            (daily_data['strike'] == current_option['strike']) & 
            (daily_data['expire_date'] == current_option['expire_date'])
        ]
        
        if options_today.empty:
            continue
            
        opt_row = options_today.iloc[0]
        dte_now = float(opt_row['dte'])
        
        # curr market val
        opt_value = (float(opt_row['c_bid']) + float(opt_row['p_bid'])) * 100
        
        # Exit
        if dte_now < EXIT_DTE:
            # clsoe ts
            cash += opt_value 
            cash += (stock_inventory * current_price) 
            
            print(f"[{date.date()}] CLOSE (Roll): PnL Check: {cash - CAPITAL:.2f}")
            
            stock_inventory = 0
            current_option = None 
            opt_row = None # Clear this so we don't calculate P&L on a closed option
            
        # hedge !!
        else:
            # Check Deltas
            net_opt_delta = (float(opt_row['c_delta']) + float(opt_row['p_delta'])) * 100
            total_delta = net_opt_delta + stock_inventory
            
            if abs(total_delta) > HEDGE_THRESHOLD:
                shares_trade = -1 * total_delta
                cash -= (shares_trade * current_price)
                stock_inventory += shares_trade

    # daily equity
    if current_option and opt_row is not None:
        c_mid = (float(opt_row['c_bid']) + float(opt_row['c_ask'])) / 2
        p_mid = (float(opt_row['p_bid']) + float(opt_row['p_ask'])) / 2
        current_opt_val = (c_mid + p_mid) * 100
    else:
        current_opt_val = 0
        
    total_equity = cash + (stock_inventory * current_price) + current_opt_val
    portfolio_value_history.append({'date': date, 'equity': total_equity})

print("Backtest Complete.")
```

```python
res_df = pd.DataFrame(portfolio_value_history)
res_df.set_index('date', inplace=True)

# Calculate Daily Returns
res_df['pct_change'] = res_df['equity'].pct_change()
res_df['drawdown'] = res_df['equity'] - res_df['equity'].cummax()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# portfolio value
ax1.plot(res_df.index, res_df['equity'], label='Gamma Scalping Strategy', color='blue')
ax1.set_title('Delta-Neutral Gamma Scalping (2022)')
ax1.set_ylabel('Portfolio Value ($)')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Drawdown
ax2.fill_between(res_df.index, res_df['drawdown'], 0, color='red', alpha=0.3)
ax2.set_title('Strategy Drawdown')
ax2.set_ylabel('Drawdown ($)')
ax2.grid(True, alpha=0.3)

plt.show()

total_return = (res_df['equity'].iloc[-1] - CAPITAL) / CAPITAL * 100
max_dd = res_df['drawdown'].min()
volatility = res_df['pct_change'].std() * np.sqrt(252) * 100

print(f"Total Return: {total_return:.2f}%")
print(f"Max Drawdown: ${max_dd:.2f}")
print(f"Annualized Volatility: {volatility:.2f}%")
```
